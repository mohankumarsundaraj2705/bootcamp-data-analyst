```python
import matplotlib.pyplot as plt
import numpy as np

def plot_happiness_score(happiness_scores, years):
    plt.figure(figsize=(8, 6))
    
    plt.scatter(happiness_scores, years, color='blue', label='Happiness Score')
    
    plt.xlabel("Happiness Score (WHR)")
    plt.ylabel("Year")
    plt.title("Happiness Score Over the Years")
    
    plt.grid(True, linestyle='--', alpha=0.6)
    
    plt.legend()
    plt.show()

happiness_scores = [5.2, 5.5, 5.8, 6.1, 6.4, 6.7, 7.0]  
years = [2015, 2016, 2017, 2018, 2019, 2020, 2021]   

plot_happiness_score(happiness_scores, years)
```
